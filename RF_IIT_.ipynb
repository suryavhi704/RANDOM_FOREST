```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## **Getting information about the Dataset:-**

```python
df=pd.read_csv('customer_churn_.csv')
df.info()
```

## **Performing EDA Operations:-**

```python
#checking for null values:-
df.isnull().sum().sum()
```

```python
#convertiong the 'TotalCharges ' column into its correct type.
df['TotalCharges']=pd.to_numeric(df['TotalCharges'],errors='coerce')
```

```python
#checking for null values again:-
df.isnull().sum().sum()
```

```python
#Droping the null values:-
df.dropna(inplace=True)
```

```python
#Checking for duplicate values:-
df.duplicated().sum()
```

```python
#performing label encoding:-
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in df.columns:
  df[i]=le.fit_transform(df[i])
```

```python
#cheking again about the info:-
df.info()
```

## **Creating RandomForest model:-**

```python
#Required libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
x=df.drop(columns='Churn')
y=df['Churn']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
rf=RandomForestClassifier(n_estimators=70,random_state=42)
rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
```

```python
#checking the accuracy
print("Accuracy of the model",accuracy_score(y_test,y_pred)*100,'%')
```

## **Trying to increase the accuracy of the model using GridSearch:-**

```python
from sklearn.model_selection import GridSearchCV
```

```python
base_model=RandomForestClassifier(random_state=42)
```

```python
param_grid={
    'n_estimators':[50,70,100,120],
    'max_depth':[2,4,6,10],
    'min_samples_split':[2,4,6,8],
    'min_samples_leaf':[2,4,6,8],
    'criterion':['gini','entropy']

}
```

```python
grid_search=GridSearchCV(estimator=base_model,param_grid=param_grid,cv=5,scoring='accuracy')
```

```python
grid_search.fit(x_train,y_train)
```

```python
print(grid_search.best_params_)
```

```python
final_model=RandomForestClassifier(criterion='entropy',max_depth=10,min_samples_leaf=6,min_samples_split=2,n_estimators=70,random_state=42)
final_model.fit(x_train,y_train)
```

```python
final_predict=final_model.predict(x_test)
```

```python
print("Accuracy of the model",accuracy_score(y_test,final_predict)*100,'%')
```
